conditinal frequency distribution: all words in the inaugural address corpus that begin with america or citizen are counted; separate counts are kept for each address; these are plotted so that trends in usage over time can be observed. counts are not normalized for document length

```python
cfd = nltk.ConditionalFreqDist(
            (target,fileid[:4])
            for fileid in inaugural.fileids()
            for w in inaugural.words(fileid)
            for target in ['america','citizen']
            if w.lower().startswith(target)
            )
cfd.plot()
```

2.2 conditonal frequency distributions

A conditional frequency distributions is a cocllection of frequency distributions, each one for a different condition

For each condition, it counds the number of words

for example we have a count of the word 'Monday' for News texts and a count of 'Monday' for Romance

Conditions and Events

A frequency distribution counts observable events, such as the appearance of words sin a text. A conditional frequency distribution needs to pair each event with a condition. Instead of processing a sequency of words, we process a sequence of pairs;

`text = ['The','Fulton','County','Grand','Jury','said',...]`

`pairs = [('news','The'),('news','Fulton'),('news','County'),...]`

each pair has the form of `(condition,event)`

### Counting Words by Genre

`FreqDist()` takea a simple list as input

`ConditonalFreqDist()` takes a list of pairs

```python
from nltk.corpus import brown

cfd = nltk.ConditionalFreqDist(
            (genre,word) for genre in brown.categories() for word in brown.words(categories=genre)
)
cfd
```

```python
genre_word = [(genre,word) for genre in ['news','romance'] for word in brown.words(categories=genre)]
len(genre_word)
```

Plotting and tabulating distributions

We can plot conditonal distributions. This code generates the pair `('america','1865')` for every instance of a word whose lower cased form starts with Americans - in the file 1865-Lincoln.txt

```python
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
        (target,fileid[:4])
        for fileid in inaugural.fileids()
        for w in inaugural.words(fileid)
        for target in ['america','citizen']
        if w.lower().startswith(target)
)
```

```python
cfd.conditions()
```

```python
cfd.tabulate(conditions=['English','German_Deutsch'],samples=range(10),cumulative=True)
```

```python
from nltk.corpus import udhr
languages = ['Chickasaw','English','German_Deutsch','Greenlandic_Inuktikut','Hungarian_Magyar','Ibibio_Efik']
cfd = nltk.ConditionalFreqDist((lang,len(word)) for lang in languages for word in udhr.words(lang+'-Latin1'))
cfd.conditions()
```

```python
cfd.tabulate(conditions=['English','German_Deutsch'],samples=range(10),cumulative=True)
```

interpretation above; column 9, row English - there are 1638 words of english to have 9 or fewer letters.

#Generating Random Text with Bigrams

We can use conditonal frequency distributions to create a table of bigrams (word pairs). The bigrams() function takes a list of words and builds a list of consecutive word pairs.

```python
sent = ['In','the','beginning','God','created','the','heaven','and','the','earth']
list(nltk.bigrams(sent))
```

```python
def generate_model(cfdist,word,num=15):
    for i in range(num):
        print(word)
        word = cfdist[word].max()
```

```python
text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
```

```python
cfd['living']
```

```python
generate_model(cfd,'living')
```

```python
cfd.plot()
```

# Lexical Resources

A lexicon, or lexical resource, is a collection of words/phrases along with associated information such as part of speech and sense definitions. Lexical resources are secondary to texts, and are usually created and enriched with the help of texts. for example if we have defined a text `my_text` then `vocab = sorted(set(my_text))` builds the vocabulary of my_text, where as `word_freq = FreqDist(my_text)`
