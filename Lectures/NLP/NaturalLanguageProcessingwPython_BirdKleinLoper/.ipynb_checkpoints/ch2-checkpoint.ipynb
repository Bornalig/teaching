```python
#find all the basic infor about each file in the guten berg docs
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print(int(num_chars/num_words),int(num_words/num_sents),int(num_words/num_vocab),fileid)
```

```python
#convenient resource for studying systematic differences between genres is known as stylistics

from nltk.corpus import brown
news_text = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can','could','may','might','must','will']
for m in modals:
    print(m+':',fdist[m])
```

```python
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
[fileid[:4] for fileid in inaugural.fileids()]
```

conditinal frequency distribution: all words in the inaugural address corpus that begin with america or citizen are counted; separate counts are kept for each address; these are plotted so that trends in usage over time can be observed. counts are not normalized for document length

```python
cfd = nltk.ConditionalFreqDist(
            (target,fileid[:4])
            for fileid in inaugural.fileids()
            for w in inaugural.words(fileid)
            for target in ['america','citizen']
            if w.lower().startswith(target)
            )
cfd.plot()
```

```python
from nltk.corpus import udhr
languages = ['Chickasaw','English','German_Deutsch','Greenlandic_Inuktikut','Hungarian_Magyar','Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
                (lang,len(word))
                for lang in languages
                for word in udhr.words(lang+'-Latin1')
)
cfd.plot(cumulative=True)
```

#common structure for text corpora: the simpleest kind of corpus is a collection of isolated texts with no particular organization. 

some corpora are structured into categories, such as genres (brow corpus), some categories overlap (reuters corpus), other corpora represent language use over time

There are many different functionalities of corpuses in NLTK

* `fileids()`
* `fileids([categorical])`
* `categories`
* `categories([fileids])`
* `raw()`
* `raw(fileids=[f1,f2,f3])`
* `raw(categories=[c1,c2])`
*  `words()`
* `words(fileids=[f1,f2,f3])`

```python
raw = gutenberg.raw('burgess-busterbrown.txt')
raw[1:20]
```

```python
words = gutenberg.words('burgess-busterbrown.txt')
words[1:20]
```

```python
sents = gutenberg.sents("burgess-busterbrown.txt")
sents[1:20]
```
