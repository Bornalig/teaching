# Problem 1

Make a class `binary_classification_report` that returns all the metrics from the lecture for both the train and test set

* It's initalization should be; `binary_classification_report(model,X_train,X_test,y_train,y_test)`
* it returns;
    * String report:
        * `accuracy_score`
        * `log_loss`
        * `f1_score`
        * `classification_report`
    * Graphic report:
        * `roc-auc` graphic
        * `confusion_matrix` graphic (normalized)
        * `auc` metric
    * return both the string and graphic report for the train set and test set in one `report()` method

# use the data & model set from lecture 9 to input into the class

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
columns = ['times pregnant','glucose_concentration','blood-pressure','skinfold-thickness',
           'serum-insulin','BMI','diabetes-function','age','class']
data = pd.read_csv('https://raw.githubusercontent.com/JoeGanser/Datasets-1/master/pima-indians-diabetes.csv',names=columns)
X = data.drop('class',axis=1)
y = data['class']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)

logreg = LogisticRegression(C=10)
logreg.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,log_loss,f1_score,roc_curve,auc,plot_confusion_matrix
import matplotlib.pyplot as plt
class binary_classification_report:
    def __init__(self,model,X_train,y_train,X_test,y_test):
        self.model = model
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        #predictions
        self.y_pred_test = self.model.predict(X_test)
        self.y_pred_proba_test = self.model.predict_proba(X_test)
        self.y_pred_train = self.model.predict(X_train)
        self.y_pred_proba_train = self.model.predict_proba(X_train)
        
    def string_report(self,y_pred,y_pred_prob,y_obs,label):
        score = accuracy_score(y_pred,y_obs)
        logloss = log_loss(y_obs,y_pred_prob)
        f1 = f1_score(y_pred,y_obs)
        report = classification_report(y_pred,y_obs)
        
        metric_text = '''
        Metrics for {0}\n
        Accuracy Score: {1}\n
        log loss: {2}\n
        f1 score: {3}\n
        classification report: \n{4}'''.format(label,score,logloss,f1,report)
        print(metric_text)
        return metric_text
        
    def graphic_reports(self,X,y_obs,y_pred,y_pred_prob,label):
        y_pred_prob = y_pred_prob[:,1]
        false_positive_rate, true_positive_rate, thresholds = roc_curve(y_obs, y_pred_prob)
        roc_auc = auc(false_positive_rate, true_positive_rate)
        plt.figure(figsize=(5,5))
        plt.title('Receiver Operating Characteristic {0} Set \n AUC: {1}'.format(label,roc_auc))
        plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],linestyle='--')
        plt.axis('tight')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plot_confusion_matrix(self.model, X, y_obs,normalize='true')  
        plt.title('Confusion matrix for train set \n (normalized)')
        plt.show()
        
    
    def report(self):
        self.string_report(self.y_pred_test,self.y_pred_proba_test,self.y_test,'test set')
        self.graphic_reports(self.X_test,self.y_test,self.y_pred_test,self.y_pred_proba_test,'test set')
        self.string_report(self.y_pred_train,self.y_pred_proba_train,self.y_train,'train set')
        self.graphic_reports(self.X_train,self.y_train,self.y_pred_train,self.y_pred_proba_train,'train set')
```

```python
cr = binary_classification_report(logreg,X_train,y_train,X_test,y_test)
cr.report()
```
