```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('HW12_data.csv')
X = df.drop('price',axis=1)
y = df['price']
```

# Problem 1:

Use a loop to run a grid search to fit the following models onto the dataset, performing a grid search on each model for possible parameters;

* `Ridge()`, `Lasso()`, `parameters = {'alpha':np.arange(0,1,0.1)} 
* `ElasticNet()`, `parameters = {'alpha':np.arange(0,1,0.1),'l1_ratio':np.arange(0,1,0.1)}`
* `SGDRegressor()`, `parameters = {'penalty':['l1','l2'],'alpha':np.arange(0,1,0.1),'l1_ratio':np.arange(0,1,0.1),'loss':['squared_loss']}`


* Make a list in the form of `[(model1,parameters1),(model2,parameters2),...]`
* At each combination, print `GridSearchCV`'s best estimator and best score

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet,SGDRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer,r2_score
import numpy as np

alpha = np.arange(0,1,0.1)
l1_ratio = np.arange(0,1,0.1)

#for Ridge and lasso
ridge_lasso_params = {'alpha':alpha}
#for elastic_net params
elastic_net_params = {'alpha':alpha,'l1_ratio':l1_ratio}
#for sgd regressor params
sgd_params = {'penalty':['l1','l2'],'alpha':np.arange(0,1,0.1),'l1_ratio':np.arange(0,1,0.1),'loss':['squared_loss']}

models = [(Ridge,ridge_lasso_params),(Lasso,ridge_lasso_params),(ElasticNet,elastic_net_params),(SGDRegressor,sgd_params)]

for m in models:
    model = m[0]()
    params = m[1]
    gs = GridSearchCV(model,param_grid=params,scoring=make_scorer(r2_score,greater_is_better=True))
    gs.fit(X,y)
    print(gs.best_estimator_)
    print(gs.best_score_)
    print('\n')
```
