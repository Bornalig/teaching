```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mu = 0
std = 1
df = 9
snd = stats.t(df=df, scale=std)
alpha=0.05
two_tail=True
if two_tail:
    alpha = alpha/2
t_critical = stats.t.ppf(alpha,df)
t_test = 1.2405
print(t_critical)
#x = np.linspace(5, 40, 100)
x = np.linspace(-4, 4, 1000)
pdf = snd.pdf(x)
plt.figure(figsize=(14,14))
plt.plot(x, pdf)
#plt.axvline(0,label='mean',color='red',linestyle=':',alpha=1)
plt.axvline(t_test,label='t_test',color='red',linestyle='--',alpha=1)
plt.xlim(-4, 4)
plt.ylim(0,0.4)
#plt.xticks([-t_critical,t_critical])
#plt.title('Gaussian Distribution', fontsize='15')
#plt.xlabel('IQ Scores', fontsize='15')
#plt.ylabel('Probability', fontsize='15')
section1 = [j for j in x if j<t_critical]# or j>-t_critical]
plt.fill_between(section1,snd.pdf(section1),color='blue')
section2 = [j for j in x if j>-t_critical]
plt.fill_between(section2,snd.pdf(section2),color='blue',label='rejection \nregion')
plt.legend(fontsize=20,loc='upper left')
#plt.savefig('example5.png')
plt.show()
```
