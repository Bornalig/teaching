```python
link = "https://www.msnbc.com/opinion/florida-gov-ron-desantis-don-t-say-gay-bill-will-n1288662"


from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from textblob import TextBlob
import requests

response = requests.get(link)
soup = BeautifulSoup(response.content,'html5lib')
text = soup.get_text(strip=True)
blob = TextBlob(text)
print('Entire sentiment: ',blob.sentiment)
for s in blob.sentences:
    print(s)
    print(s.sentiment)
    print(s.noun_phrases)
    print('\n')
```

# Problem 12.4

```python
link = "https://www.msnbc.com/opinion/florida-gov-ron-desantis-don-t-say-gay-bill-will-n1288662"


from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import requests

response = requests.get(link)
soup = BeautifulSoup(response.content,'html5lib')
text = soup.get_text(strip=True)
blob = TextBlob(text,analyzer=NaiveBayesAnalyzer())
print('Entire sentiment: ',blob.sentiment)

for s in blob.sentences:
    print(s)
    print(s.sentiment)
    print(s.noun_phrases)
    print('\n')
```

# Problem 12.6

```python
import urllib.request
from textblob import TextBlob
from nltk.corpus import stopwords
from operator import itemgetter
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://www.gutenberg.org/files/1112/1112.txt'
text = ''
for line in urllib.request.urlopen(url):
    text = text + line.decode('utf-8')+'\n'
    
blob = TextBlob(text)
wordcloud = WordCloud(colormap='prism')
wordcloud = wordcloud.generate(text)
wordcloud.to_file('problem12-6wordcloud.png')

items = blob.word_counts.items()
items = [item for item in items if item[0] not in stop_words]
sorted_items = sorted(items,key=itemgetter(1),reverse=True)
top20 = sorted_items[1:21]
df = pd.DataFrame(top20,columns=['word','count'])
axes = df.plot.bar(y='count',x='word',legend=False)
plt.title('Top 20 word counts, romeo & juliet')
plt.gcf().tight_layout()
```
