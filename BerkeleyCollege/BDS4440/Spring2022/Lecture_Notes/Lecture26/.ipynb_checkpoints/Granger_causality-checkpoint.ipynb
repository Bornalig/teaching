```python
import pandas as pd
#define URL where dataset is located
url = "https://raw.githubusercontent.com/Statology/Miscellaneous/main/chicken_egg.txt"
#read in dataset as pandas DataFrame
df = pd.read_csv(url, sep="  ").set_index('year')
#view first five rows of DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.title('chickens in the USA')
df['chicken'].plot()
plt.subplot(1,2,2)
plt.title('Eggs layed in the USA')
df['egg'].plot()
plt.subplots_adjust(wspace=0.3)
plt.show()
```

# We must test to see if these time series are stationary to test for causality

```python
from statsmodels.tsa.stattools import adfuller 
def ADF_test(data):
    adf = adfuller(data)
    test_statisic,p_value,n_lags,n_observations,d,n = list(adf)
    print("\nStatistics analysis\n")
    print("Statistic Test : " , test_statisic)
    print("p-value : " , p_value)
    print("# n_lags : " , n_lags)
    print("No of observation: " , n_observations)
    for key,value in adf[4].items():
        print(f" critical value {key} : {value}") 
    if p_value<0.1 and (test_statisic<adf[4]['10%']):
        print('Reject Null')
    
    elif p_value<0.05 and (test_statisic<adf[4]['5%']):
        print('Reject Null')
        
    elif p_value<0.01 and (test_statisic<adf[4]['1%']):
        print('Reject Null')
    else:
        print('Fail to Reject Null Hypothesis')
```

```python
ADF_test(df['chicken'])
ADF_test(df['egg'])
```

# transform the data set using the `.diff().dropna() methods`

```python
dff = df.diff().dropna()
ADF_test(dff['chicken'])
ADF_test(dff['egg'])
dff.head()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Transformed chickens in the USA')
dff['chicken'].plot()
plt.subplot(1,2,2)
plt.title('Transformed Eggs layed in the USA')
dff['egg'].plot()
plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
from scipy.stats import f
def granger_test(data,maxlag,alpha):
    gt = grangercausalitytests(data,maxlag)
    for i in range(1,maxlag):
        print(gt[i][0]['params_ftest'])
        F_stat,p_value = gt[i][0]['params_ftest'][:2]
        dfd,dfn = gt[i][0]['params_ftest'][-2:]
        F_critical = f.ppf(1-alpha,dfn,dfd)
        print('Lag {}'.format(i))
        if F_stat>=F_critical and p_value<=alpha:
            print('Reject Null hypothesis')
        else:
            print('Fail to reject null hypothesis')
        print('\n')
```

```python
print(maxlag)
alpha=0.01
for index,data in enumerate([pd.concat([dff['egg'],dff['chicken']],axis=1),pd.concat([dff['chicken'],dff['egg']],axis=1)]):
    for i in range(1,10):
        maxlag = i
        if index==0:
            print('egg - > chicken')
        else:
            print('chicken -> egg')
        granger_test(data,maxlag,alpha)
```
