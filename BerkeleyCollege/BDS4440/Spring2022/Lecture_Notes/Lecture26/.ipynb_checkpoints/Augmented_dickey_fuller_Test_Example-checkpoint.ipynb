* We can get the moving avg price over the last 30 days using the `.rollling(30).mean()` method on the `Close` column.
* We can find the difference between today's price and yesterdays price using the `.shift(1)` pandas method.
* We can find the voliatility of the noise by finding the moving standard deviations using the `daily_difference` column

```python
n=30
df['{}_day_mean'.format(n)] = df['Close'].rolling(n).mean()
df['daily_difference'] = df['Close'] - df['Close'].shift(1)
df['{}_day_volatility'.format(n)] = df['daily_difference'].rolling(n).std()
df.tail()
```

Lets look at the closing prices seasonal value. This may be easier if we resample the data weekly, getting the sum of closing prices over the  week

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose  = seasonal_decompose(df['Close'].resample('W').sum(),freq=52)
fig = plt.figure()
fig = decompose.plot()
fig.set_size_inches(12,8)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller 

def ADF_test(data):
    adf = adfuller(data)
    test_statisic = float(adf[0])
    p_value = float(adf[1])
    n_lags = float(adf[2])
    n_observations = float(adf[3])
    print("\nStatistics analysis\n")
    print("Statistic Test : " , test_statisic)
    print("p-value : " , p_value)
    print("# n_lags : " , n_lags)
    print("No of observation: " , n_observations)
    for key,value in adf[4].items():
        print(f" critical value {key} : {value}")
        
    if p_value<0.1 and (test_statisic<adf[4]['10%']):
        print('Reject Null')
    
    elif p_value<0.05 and (test_statisic<adf[4]['5%']):
        print('Reject Null')
        
    elif p_value<0.01 and (test_statisic<adf[4]['1%']):
        print('Reject Null')
    else:
        print('Fail to Reject Null Hypothesis')
```

# Stationary format: `np.log(btc) - np.log(btc).shift(1)`

```python
import numpy as np
btc = df['Close']

inverted = np.log(btc) - np.log(btc).shift(1)
ADF_test(inverted.dropna())
```
