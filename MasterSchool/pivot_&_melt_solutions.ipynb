```python
import pandas as pd

data = {'name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
        'category': ['A', 'A', 'B', 'B', 'B', 'A'],
        'value': [10, 20, 30, 40, 50, 60]}
df = pd.DataFrame(data)
df
```

`df.pivot()` returns data as it is identically, and if there are multiple values for the same combintion or row/column it returns a `ValueError`.

```python
df.pivot(index='name', columns='category', values='value')
```

```python
pd.pivot_table(df, values='value', index='name', columns='category', aggfunc='sum')
```

```python
pd.pivot_table(df, values='value', columns='category', aggfunc='sum')
```

```python
pd.pivot_table(df, values='value', index='name', aggfunc='sum')
```
